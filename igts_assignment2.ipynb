```python
#question3

matrix=[
    [(1,1),(-1,2),(5,0),(1,1)],
    [(2,3),(1,2),(3,0),(5,1)],
    [(1,1),(0,5),(1,7),(0,1)]
]


def iesds(game):
    m, n = len(game), len(game[0])
    
    surviving_rows = list(range(m))
    surviving_cols = list(range(n))
    
    eliminated_any = True
    
    print("Starting IESDS Elimination Process:")

    
    while eliminated_any:
        eliminated_any = False
        
        # finding dominated row
        
        if len(surviving_rows) > 1:
            for r1 in surviving_rows:
                for r2 in surviving_rows:
                    if r1 == r2:
                        continue
                    
                    is_dominated = True
                    for c in surviving_cols:
                        if game[r1][c][0] >= game[r2][c][0]:
                            is_dominated = False
                            break
                    
                    if is_dominated:
                        surviving_rows.remove(r1)
                        eliminated_any = True
                        break 
                if eliminated_any:
                    break 
            if eliminated_any:
                continue 
                
        # finding dominated column
        if len(surviving_cols) > 1:
            for c1 in surviving_cols:
                for c2 in surviving_cols:
                    if c1 == c2:
                        continue
                    
                    is_dominated = True
                    for r in surviving_rows:
                        if game[r][c1][1] >= game[r][c2][1]:
                            is_dominated = False
                            break
                    
                    if is_dominated:
                        surviving_cols.remove(c1)
                        eliminated_any = True
                        break
                if eliminated_any:
                    break
            if eliminated_any:
                continue

    
    print("Surviving Rows :", surviving_rows)
    print("Surviving Columns :", surviving_cols)
    
    # Check if a single cell remains
    if len(surviving_rows) == 1 and len(surviving_cols) == 1:
        r, c = surviving_rows[0], surviving_cols[0]
        print(f"✅ Predicted Outcome Declared: Cell ({r}, {c}) with payoffs {game[r][c]}")
    else:
        print("Multiple strategies survived. No single predictable outcome.")
        
    return surviving_rows, surviving_cols

iesds(matrix)
```

```python
#question4
game=[
    [(4,4),(0,0),(1,3)],
    [(0,0),(5,5),(0,0)],
    [(3,1),(0,0),(6,6)]
]

#dominant strategy for 1

cols=len(game[0])
rows=len(game)

p1best=[]
for i in range(cols):
    rowvals=[]
    for j in range(rows):
        rowvals.append(game[j][i][0])
    maxval=max(rowvals)
    maxrow=rowvals.index(maxval)
    p1best.append((maxrow,i))

p2best=[]
for i in range(rows):
    colvals=[]
    for j in range(cols):
        colvals.append(game[i][j][1])
    maxval=max(colvals)
    maxcol=colvals.index(maxval)
    p2best.append((i,maxcol))

flag=False
for i in p1best:
    if i in p2best:
        print("Nash equilibrium at ",i) 
        flag=True

if not flag:
    print("No Nash equilibrium found")
```
