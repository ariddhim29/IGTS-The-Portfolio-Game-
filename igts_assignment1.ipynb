```python
#question_2
x=str(input("enter string"))
new= x.lower()

final= "".join(new.split())
length= len(final)

i=0
half=int(length/2)
flag = True
while i < half:
    if final[i] != final [length-i-1]:
        flag= False
        break
    i=i+1

if flag == True :
    print("palindrome")
else:
    print("not palindrome")
            
```

```python
#question_7
import numpy as np
import matplotlib.pyplot as plt

#a
returns= np.random.normal(loc=0.0008, scale=0.015, size=252)
price=100

prices=[]

for i in returns:
    price=price* (1+i)
    prices.append(price)

#b
plt.plot(prices)
plt.title("Cumulative Price Series")
plt.xlabel("Day")
plt.ylabel("Prices")
plt.show()

#c
print("final price: ", prices[251])
running_max=np.maximum.accumulate(prices)

print("peak price: ", running_max[251])

drawdowns= (prices- running_max)/running_max

max_dd=np.min(drawdowns)

print("max drawdown: ", max_dd)
```

```python
#question_8
import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May','Jun',
'Jul','Aug','Sep','Oct','Nov','Dec']
year1 = [42, 45, 50, 48, 55, 60, 58, 63, 67, 70, 75, 80]
year2 = [52, 49, 58, 61, 66, 71, 68, 74, 79, 83, 88, 95]

#a
plt.plot(months, year1, label="year1")
plt.plot(months, year2, label="year2")

plt.xlabel("months")
plt.ylabel("revenue in lakhs")
plt.title("monthly revenue data (in lakhs) across 2 years for Company A")
plt.legend()
plt.show()

#b
diff=[]
diffmod=[]
for i in range(12):
    dif= year2[i]-year1[i]
    diff.append(dif)
    diffmod.append(abs(dif))


colors=[]
for i in diff:
    if (i>0):
        colors.append("green")
    else:
        colors.append("red")

plt.bar(months, diffmod, color=colors)
plt.show()
```

```python
#question_9
import numpy as np, pandas as pd
df = pd.DataFrame({
'Day': range(1, 11),
'Price': [100, 103, np.nan, 108, 107, np.nan, 112, 115, np.nan, 120],
'Volume': [500, 520, 510, np.nan, 530, 545, np.nan, 560, 570, 580]
})

prices=df["Price"].tolist()
volumes=df["Volume"].tolist()

nullp=0
nullv=0

for i in prices:
    if np.isnan(i):
        nullp = nullp+1
print("Missing prices: ",nullp)


for i in volumes:
    if np.isnan(i):
        nullv = nullv+1
print("Missing volumes: ",nullv)

df["Price"]=df["Price"].ffill()

meanv=df["Volume"].mean()
df["Volume"]=df["Volume"].fillna(meanv)

prices=df["Price"].tolist()
returns=[0]

i=1
while i>0 and i<10:
    ret= (prices[i]-prices[i-1])/prices[i-1] *100
    returns.append(ret)
    i=i+1


df["Return"]= returns

print(df)
```
